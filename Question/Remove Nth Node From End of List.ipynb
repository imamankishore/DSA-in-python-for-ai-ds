## Given the head of a linked list, remove the nth node from the end of the list and return its head.

### Input: head = [1,2,3,4,5], n = 2
Output: [1,2,3,5]


Example 2:

Input: head = [1], n = 1
Output: []


Example 3:

Input: head = [1,2], n = 1
Output: [1]
 

Constraints:

The number of nodes in the list is sz.
1 <= sz <= 30
0 <= Node.val <= 100
1 <= n <= sz

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head =  None

    def is_empty(self):
        return self.head is None

    def length (self):
        if self.is_empty():
            return 0
        else:
            curr = self.head
            count = 0
            while curr:
                count +=1
                curr = curr.next
            return count

    
    def display(self):
        if self.is_empty():
            print('Linked List is Empty')
            return
        else :
            curr = self.head
            while curr:
                print(curr.data , end = ' ')
                curr=curr.next
            print()
        

    def add_element(self , value):
        node = Node(value)
        if self.is_empty():
            self.head = node
            node.next = None
        else:
            node.next = self.head
            self.head = node
        self.display()



    def remove_nth_element_back(self, n):
        if self.is_empty():
            print('Linked List is Empty')
        else:
            m = self.length() - n
            if m == 0:
                print("Can't remove items")
            elif m == self.length():
                curr = self.head
                self.head = curr.next
                curr = None
            else:
                p = self.head
                q = None
                for _ in range(m):
                    q = p
                    p = p.next
                q.next = p.next
                p = None
            self.display()
```

```python
LL = LinkedList()
```

```python
LL.add_element(56)
LL.add_element(78)
LL.add_element(44)
LL.add_element(15)
```

```python
LL.display()
```

```python
LL.remove_nth_element_back(3)
```
