```python
from collections import deque
```

```python
def get_last_person(n , k):
    if n == 1:
        return n

    #populating our queue
    q = deque()
    for x in range(1 , n+1):
        q.append(x)


    while True:
        #removing the first k
        for _ in range(k):
            q.popleft()
            if len(q) == 1:
                return q[0]
        #removing from last k
        for _ in range(k):
            q.pop()
            if len(q) == 1:
                return q[0]
```

```python
get_last_person(9 , 3)
```
