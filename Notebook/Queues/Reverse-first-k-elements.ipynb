```python
from collections import deque
```

```python
q1 = deque([1,2,3,4,5])
q2 = deque([4,3,2,1])
```

```python
q2
```

```python
def reverse_stack_k(q , k):
    stack = deque()
    result = deque()

    for _ in range(k):
        stack.append(q.popleft())

    while stack:
        result.append(stack.pop())
        
    while q:
        result.append(q.popleft())

    return result
```

```python
reverse_stack_k(q1 , 3)
```

```python
reverse_stack_k(q2,2)
```
