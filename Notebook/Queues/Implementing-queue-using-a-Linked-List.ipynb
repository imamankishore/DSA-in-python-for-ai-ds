```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class QueueUsingLinkedList:
    def __init__(self):
        self.front = None
        self.rear = None
    def is_empty(self):
        return self.front is None


    def display(self):
        if self.is_empty():
            print('Empty Queue')
        else:
            curr = self.front
            while curr:
                print(curr.data , end=' ')
                curr = curr.next
            print()


    def length(self):
        if self.is_empty():
            return 0
        else:
            n = 0
            curr = self.front
            while curr:
                n += 1
                curr = curr.next
                return n


    def enqueue(self , data):
        node = Node(data)
        if self.is_empty():
            self.front = node
            self.rear = node
        else:
            self.rear.next = node
            self.rear = node
            node.next = None
        self.display()



    def dequeue(self):
        if self.is_empty():
            print("Cannot delete as queue is empty")
        elif self.length==1:
            self.front = None
            self.rear = None
        else:
            curr = self.front
            self.front = curr.next
            curr = None
        self.display()
```

```python
que = QueueUsingLinkedList()
```

```python
que.is_empty()
```

```python
que.display()
```

```python
que.length()
```

```python
que.enqueue(56)
que.enqueue(30)
que.enqueue(63)
que.enqueue(45)
que.enqueue(78)
```

```python
que.dequeue()
```
