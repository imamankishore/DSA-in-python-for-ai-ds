## Implementation of queue using a Linked Listm

```python
class QueueUsingList:
    def __init__(self):
        self.queue = []


    def is_empty(self):
        return len(self.queue) == 0


    def display(self):
        if self.is_empty():
            print('Cannot Print the empty queue')
        else:
            print(self.queue)



    def enqueue(self, data):
        self.queue.append(data)
        self.display()



    def dequeue(self):
        self.queue.pop(0)
        self.display()

    def front_peek(self):
        print(self.queue[0])


    def rear_peek(self):
        print(self.queue[-1])


```

```python
Queue = QueueUsingList()
```

```python
Queue.enqueue(56)
Queue.enqueue(89)
Queue.enqueue(5)
Queue.enqueue(36)
Queue.enqueue(78)
Queue.enqueue(1)
```

```python
Queue.dequeue()
```

```python
Queue.dequeue()
```

```python
Queue.front_peek()
```

```python
Queue.rear_peek()
```
