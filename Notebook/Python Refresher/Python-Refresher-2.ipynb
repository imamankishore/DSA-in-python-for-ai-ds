#### 2.1 If - else

```python
a = 7 
if a > 5:
    print('Greater then 5')
else :
    print('Not greater than 5')
```

```python
## checking the number is even or odd

a = int(input('Enter the number '))

if (a%2 == 2):
    print('The number is even')
else :
    print('The number is odd')
```

#### 2.2 if-elif-else

```python
marks = 76

if marks > 90:
    print('Excellent')
elif marks >75:
    print('Very good')
elif marks >50:
    print('Good')
else :
    print('bad')
```
