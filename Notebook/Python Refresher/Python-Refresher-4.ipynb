```python
names = ['bruce' , 'Peter', 'Barry', 'Steve']
for i , x in enumerate(names):
    print(i ,x)
```

```python
names
```

```python
colors = ['red', 'blue' , 'orange' , 'pink']
ages = [23,20,19,18,15,36,48]
```

```python
for name , age , color in zip(names , ages , colors):
    print(f"{name} is {age} years old and like {color} color")
```
