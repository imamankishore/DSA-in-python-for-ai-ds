# Classes

- Naming convention : camel-case
- Define constructor: __init__
- Attributes
- Methods
- Class Objects

```python
class MyRectangle :                          # camel-case naming convention
    def __init__(self , length , breadth):   # init
        self.length = length                 # attributes
        self.breadth = breadth

    def get_perimeter(self):
        return 2*(self.length + self.breadth)

    def get_area(self):
        return self.length * self.breadth
```

```python
obj1 = MyRectangle(length = 5 , breadth = 8)
```

```python
obj1.get_area()
```

```python
obj1.get_perimeter()
```
