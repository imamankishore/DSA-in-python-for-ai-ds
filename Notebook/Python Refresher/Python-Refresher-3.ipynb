#### 1. For Loop

```python
l1 = [1,2,3,4,5]

for i in l1:
    print(i)
```

```python
for i in range(20,31):
    if (i%2 == 0):
        print(i)
```

```python
for i in range(20,31 ,2):
    print(i)
```

#### 3.2 While loop

```python
a = 1 

while a <= 5:
    print(a)
    a +=1
```
