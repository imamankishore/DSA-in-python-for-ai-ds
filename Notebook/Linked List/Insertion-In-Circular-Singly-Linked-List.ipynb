- Case2. List is not empty

# Insertion at end in Circular singly Linked List

- Case1. List is empty
- Case2. List is not empty

# Insertion at middle in Circular singly Linked List

- Case1. - Invalid Location
- Case2. - At Begining
- Case3. - At End
- Case4. - At middle

```python
class Node:
    def __init__ (self,data):
        self.data = data
        self.next = None
```

```python
class CricularSinglyLinkedList:
    def __init__ (self):
        self.head = None
        self.length = 0


    # To check for Empty List
    def is_empty(self):
        return self.head is None



    # To display
    def display(self):
        if self.is_empty():
            print('Empty List')
        else:
            current = self.head
            while current != None :                #While True:
                print(current.data , end=" ")
                current = current.next
                if current == self.head:
                    break
            print()



    # To insert at begining
    def insertion_at_begining(self , value):
        node = Node(value)
        if self.is_empty():
            self.head = node
            node.next = self.head
        else :
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = node
            node.next = self.head
            self.head = node
        self.length +=1
        self.display()




    # To insert at end
    def insertion_at_end(self , value):
        node = Node(value)
        if self.is_empty():
            self.head = node
            node.next = self.head
        else:
            current = self.head
            while current.next!= self.head:
                current = current.next
            current.next = node
            node.next = self.head
        self.length +=1
        self.display()




    # To insert at middle
    def insert_at_middle(self ,pos , value):
        n = self.length
        if (pos<0) or (pos >n):
            print('Envalid postion or this position ndoesnt exisit')
        elif pos == 0:
            self.insertion_at_begining(value)
        elif pos == n:
            self.insertion_at_end(value)
        else:
            node = Node(value)
            p = self.head
            q = None
            for _ in range(pos):
                q = p
                p = p.next
            q.next = node
            node.next = p
            self.length +=1
            self.display()
```
