```python
from collections import deque
```

```python
def revrse_string_using_stack(string):
    stack = deque()
    result =''


    #push the character in the stack
    for ch in string:
        stack.append(ch)

    #pop character from stack
    while stack:
        popped = stack.pop()
        result += popped

    return result
```

```python
revrse_string_using_stack('Happy')
```
