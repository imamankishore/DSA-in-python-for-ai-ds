## Implementation of Stack using `LIST`

```python
class MyStack:
    def __init__ (self):
        self.stack = []


    def get_length(self):
        return len(self.stack)


    def is_empty(self):
        return self.get_length() == 0


    def display(self):
        if self.is_empty():
            print('Empty Stack')
            return
        else:
            for i in self.stack[::-1]:
                print(i)
            print()



    def push(self,data):
        self.stack.append(data)
        self.display()



    def pop(self):
        if self.is_empty():
            print('Empty Stack')
            return None
        else:
            popped = self.stack.pop()
            self.display()
            print(f'Popped element : {popped}')



    def peek(self):
        if self.is_empty():
            print('Empty Stack')
            return None
        else :
            print(self.stack[-1])
```

```python
stack = MyStack()
```

```python
stack.get_length()
```

```python
stack.display()
```

```python
stack.push(56)
stack.push(45)
stack.push(78)
```

```python
stack.peek()
```
