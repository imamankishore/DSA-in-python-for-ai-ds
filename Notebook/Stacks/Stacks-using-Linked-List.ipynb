## Implemention using `Linked List`

```python
class Node:
    def __init__(self , data):
        self.data = data
        self.next = None
```

```python
class StackUsingLinkedList:
    def __init__ (self):
        self.top = None

    def is_empty(self):
        return self.top is None


    def length(self):
        pos = 0
        curr = self.top
        while curr:
            pos+=1
            curr = curr.next
        return pos

        
    def display(self):
        if self.is_empty():
            print('Empty list')
        else:
            curr = self.top
            while curr:
                print(curr.data)
                curr = curr.next
            print()


    def push(self , value):
        node = Node(value)
        if self.is_empty():
            self.top = node
            node.next = None
        else:
            node.next = self.top
            self.top = node
        self.display()

    
    def pop(self):
        if self.is_empty():
            print('Cannot delete from an empty stack')
            return
        elif self.length ==1:
            self.top = None
            print("Stack is Empty")
        else :
            curr = self.top
            self.top = curr.next
            curr = None

        self.display()


    def peek(self):
        print(self.top.data)

```

```python
my_stack = StackUsingLinkedList()
```

```python
my_stack.push(56)
```
