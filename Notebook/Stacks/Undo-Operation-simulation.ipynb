```python
from collections import deque
```

```python
def undo_operation(commands):
    stack = deque()


    for command in commands:
        if command == 'undo' :
            if stack:
                stack.pop()
            else:
                print('Empty stack cannot perform the undo operation')
        else:
            stack.append(command[-1])
    return ''.join(stack)
```

```python
commands = [
    'type A',
    'type B',
    'undo',
    'type C'
]

res = undo_operation(commands)
```
