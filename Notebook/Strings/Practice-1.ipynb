### Count Frequency of Characters

- assume all characters are lower 
- string may include special character

```python
s ='__hel lo 123#world'


vowels = 0
consonents = 0
 
for ch in s:
    if ch.isalpha():
        if ch in 'aeiou':
            vowels +=1
        else:
            consonents += 1
print(f"There are {vowels} vowels and {consonents} consonents")
```
