# Palindrome String

```python
s ='madam'


s_rev= s[::-1]

if s == s_rev:
    print("Its a Palindrome!")
else:
    print("Its not a palindrome!")
```

# Check for Anagrams

- Returns True if two strings contain the same characters in any order
- Strings will be of same length and lowercase

### Solving by time complexity n-logn

```python
s1 = 'listen'
s2 = 'silent'

s1_sorted = sorted(s1)
s2_sorted = sorted(s2)

if len(s1_sorted) == len(s2_sorted):
    if s1_sorted == s2_sorted:
        print('Its a anagram!')
else :
    print('Length of the strings are not same!')
```

- sorting - logn
- comparision - n
- Time - complexicity - nlogn

### Solving by time complexity n

```python
mapping = [0] *26


for ch in s1:
    index = ord(ch) - 97
    # print(ch , ord(ch) , index)
    mapping[index] += 1
# mapping
for ch in s2:
     index = ord(ch) - 97
     mapping[index] -=1
     # print(ch , ord(ch) , index)
mapping
```

- space fixed and constant so space complexicity is O(1)
