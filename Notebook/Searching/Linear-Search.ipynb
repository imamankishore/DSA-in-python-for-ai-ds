```python
def linear_search(arr , target):
    for index , element in enumerate(arr):       # enumerate gives the index while traversing so we are storing in index
        if element == target :
            return index
    return -1
```

```python
data = [4,5,6,9,8,7,2]
```

```python
linear_search(data , 9)
```

```python
linear_search(data , 45)
```
