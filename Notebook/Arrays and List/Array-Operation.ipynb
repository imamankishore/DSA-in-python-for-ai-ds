# Creating Array via Numpy

```python
import numpy as np
```

```python
np1  = np.array([10,20,30,40,50])
np2 = np.array([40,50,10,20,30])
```

```python
np1[1]
```

```python
for i in np1:
    print(i , end= " ")
```

```python
np.insert(np1,3,35)
```

```python
np.delete(np1, 3)
```

```python
np1[2]= 35
np1
```

```python
np3 = np.sort(np2)
```

```python
np3
```

```python
np3 = np1 + np2
```

```python
np3 = np.concatenate((np1,np2))
```
