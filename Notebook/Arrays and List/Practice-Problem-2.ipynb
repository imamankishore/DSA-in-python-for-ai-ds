# Remove Duplicates from a List

```python
numbers = [1,2,3,4,8,4,1,7]

set(numbers)
```

```python
list(set(numbers))
```

```python
unique_elements = []

for i in numbers:
    if i not in unique_elements:
        unique_elements.append(i)
print(unique_elements)
```

# Remove Duplicates from a Array

```python
import numpy as np

arr1 = np.array([1, 3, 4, 8, 3, 7])

arr2 = np.unique(arr1)

print(arr2)
```

# Reverse a List(by writing a algorithm)

```python
data = [1,3,6,9,7,8,1,4]

data[::-1]                        # a trick
```

```python
data = [1,3,6,9,7,8,1,4]


n = len(data)
reversed_data=[]


for i in range(n-1,-1,-1):
    reversed_data.append(data[i])
print(reversed_data)
```
