# Finding the Largest Numbe in a Given List

```python
numbers = [12,45,7,89,34,87,99,102]

largest_num=0

for i in numbers:
    if (i>largest_num):
        largest_num = i
print(f"{largest_num} is the largest number")
```

# Finding the Largest Number in a Given Array

```python
import numpy as np

arr1 = np.array([1,5,6,7,8,9,77,89])
```

```python
largest_num=0

for i in arr1:
    if( i > largest_num):
        largest_num=i
print(largest_num)
```

# Finding the Largest Number in a Given 2D List or a matrix of 3X3

```python
matrix  = [
    [1,2,3],
    [88,45,79],
    [56,89,91]
]


m = len(matrix)
n = len(matrix[0])


largest_num =0 

for i in range(m):
    for j in range(n):
        if(matrix[i][j]>largest_num):
            largest_num=matrix[i][j]
print(largest_num)
```

# Count the occurences of the list

```python
items = [1,8,9,9,6,4,7,1,7,1,4,4,1,6,7,8,7,4,7,6,7]
target = int(input("Enter the target value"))
count = 0 


for i in items:
    if (i == target):
        count+=1
print(count)
```

# Count the occurences of the Matrix of 3X3

```python
matrix1 = [
    [1,5,1],
    [4,5,1],
    [1,2,1]
]



target = 5
count = 0
m = len(matrix1)
n = len(matrix1[0])


for i in range(m):
    for j in range(n):
        if (matrix1[i][j] == target):
            count+= 1
print(count)
```
