# 2D arrat-or-list-creation

### creation

```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9,]
]
```

```python
type(matrix)
```

- type is list but its structure is 2D

### display

```python
for row in matrix:
    print(row)
```

```python
def display_matrix(matrix):
    for row in matrix:
        print (row)
```

```python
display_matrix(matrix)
```

### traversal

```python
for row in matrix:
    for i in row:
        print(i , end=" ")
    print()
```

```python
m = len(matrix)            # m = row
n = len(matrix[0])      # n = columns
 
for i in range (m):       
    for j in range (n):
        print(matrix[i][j] , end= " ")
    print()
```
